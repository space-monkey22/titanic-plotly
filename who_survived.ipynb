```python
from numpy import double
import pandas as pd
import plotly.express as px
```

```python
df = pd.read_csv("train.csv")
```

```python
df.head()
```

```python
import dash
from dash import dcc, html
import dash_bootstrap_components as dbc
import plotly.express as px


# Pie chart for Passenger in to Embarked %
pie_chart = px.pie(names=df["Embarked"].unique(), values=df.groupby("Embarked")["Embarked"].count(), hole=0.5)

# Box plot Based on Pclass vs Age
box_plot = px.box(df, y="Age", x="Pclass")

# Survival Histogram based on Pclass
group_plot = px.histogram(df, x="Survived", color="Pclass", barmode='group')

# Survival Histogram Based on Sex
group_plot1 = px.histogram(df, x="Survived", color="Sex", barmode='group')
# Histogram of SibSp by Sex
sibsp_plot = px.histogram(df, x="SibSp", color="Sex", barmode='group')

# Survival Rate Pie Chart
pie_chart1 = px.pie(names=df["Survived"].unique(), values=df.groupby("Survived")["Survived"].count(), hole=0.5)


# Line plot
line_plot = px.line(df, x="Fare", y=["Fare_Tax", "Luggage Charges", "Food Charges"])

# Dash app setup with Bootstrap stylesheet
app = dash.Dash(__name__, external_stylesheets=[dbc.themes.BOOTSTRAP])

def create_card(title, content, color):
    return dbc.Card(
        dbc.CardBody([
            html.H4(title, className="card-title text-center"),
            html.Hr(),
            content
        ]),
        color=color, inverse=True, className="mb-4"
    )

# Layout with three columns using Bootstrap
app.layout = dbc.Container([
    dbc.Row([
        dbc.Col(create_card(
            "Passenger Deaths by Class",
            html.Div([
                html.H5(f"First class Passenger Deaths: {len(df.loc[(df['Pclass'] == 1) & (df['Survived'] == 0)])}"),
                html.H5(f"Second class Passenger Deaths: {len(df.loc[(df['Pclass'] == 2) & (df['Survived'] == 0)])}"),
                html.H5(f"Third class Passenger Deaths: {len(df.loc[(df['Pclass'] == 3) & (df['Survived'] == 0)])}")
            ]),
            color="danger"
        ), width=4),

        dbc.Col(create_card(
            "Passenger Survival by Class",
            html.Div([
                html.H5(f"First class Passenger Survived: {len(df.loc[(df['Pclass'] == 1) & (df['Survived'] == 1)])}"),
                html.H5(f"Second class Passenger Survived: {len(df.loc[(df['Pclass'] == 2) & (df['Survived'] == 1)])}"),
                html.H5(f"Third class Passenger Survived: {len(df.loc[(df['Pclass'] == 3) & (df['Survived'] == 1)])}")
            ]),
            color="success"
        ), width=4),

        dbc.Col(create_card(
            "Fare & Charges",
            html.Div([
                html.H5(f"Average Fare Value: {int(df['Fare'].mean())}"),
                html.H5(f"Average Fare Tax: {int(df['Fare_Tax'].mean())}"),
                html.H5(f"Average Luggage Charges: {int(df['Luggage Charges'].mean())}")
            ]),
            color="info"
        ), width=4),
    ]),

    html.Hr(),

    dbc.Row([
        dbc.Col(dbc.Card(
            dcc.Graph(figure=pie_chart), body=True, color="light", className="shadow-sm"
        ), width=4),

        dbc.Col(dbc.Card(
            dcc.Graph(figure=group_plot), body=True, color="light", className="shadow-sm"
        ), width=4),

        dbc.Col(dbc.Card(
            dcc.Graph(figure=box_plot), body=True, color="light", className="shadow-sm"
        ), width=4),
    ]),

    html.Hr(),

    dbc.Row([
        dbc.Col(dbc.Card(
            dcc.Graph(figure=line_plot), body=True, color="light", className="shadow-sm"
        ), width=4),

        dbc.Col(dbc.Card(
            dcc.Graph(figure=group_plot1), body=True, color="light", className="shadow-sm"
        ), width=4),

        dbc.Col(dbc.Card(
            dcc.Graph(figure=pie_chart1), body=True, color="light", className="shadow-sm"
        ), width=4),
    ]),

    html.Hr(),

    # Histogram of Sibsp
    dbc.Row([
        dbc.Col(dbc.Card(
            dcc.Graph(figure=sibsp_plot), body=True, color="light", className="shadow-sm"
        ), width=12)
    ])
], fluid=True)

if __name__ == '__main__':
    app.run_server(debug=True)
```
